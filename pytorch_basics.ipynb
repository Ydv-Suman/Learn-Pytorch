## Step 1 Data

-- spliting data into training and testing data set

```python
weight = 0.9
bias = 0.5

start = 0
end = 20
step = 0.5

x = torch.arange(start, end, step)
y = weight * x + bias
```

```python
x[:5], y[:5]
```

```python
len(x), len(y)
```

```python
split = int(0.8 * len(x))
x_train, y_train = x[:split], y[:split]
x_test, y_test = x[split:], y[split:]

len(x_train), len(y_train), len(x_test), len(y_test)
```

```python
## plot prediction
import matplotlib.pyplot as plt

def plot_prediction(train_data=x_train, train_labels=y_train,
                    test_data=x_test, test_labels=y_test,
                    prediction=None):
    plt.figure(figsize=(10,8))
    plt.scatter(train_data, train_labels, c="green", label="Training")
    plt.scatter(test_data, test_labels, c="blue", label="Testing")

    if prediction is not None:
        plt.scatter(test_data, prediction, c="red", label="Prediction")
        plt.legend(prop={"size":14})
    else:
        plt.legend(prop={"size":14})

    plt.show()
```

## Step 2: Build Method

## Algorithms:
* Gradient descent
* Backpropogation

```python
# Create a linear regression model class
class LinearRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.weight = nn.Parameter(torch.rand(1, requires_grad=True, dtype=torch.float))
        self.bias = nn.Parameter(torch.rand(1, requires_grad=True, dtype=torch.float))

    # forward method defines the computation in the model
    def forward(self, x: torch.Tensor) -> torch.Tensor:   # x is input data
        return self.weight * x + self.bias
```

## pytorch mdoel building essentials
* torch.nn - cotains all of the building for computional graphs(a neural netwok can be considered a computional graph)
* torch.nm.Parameters  -  what parameters should  our model tey and often a pyTorch layer from torch.n will set these for us
* torch.nn.Module - The base class for all neural network modukes, if you subclass it, you should overwrite forward()
* torch.optim -this where the optimizers in pyTorch live, they help with gradient descent


```python
# Create a random seed
torch.manual_seed(42)

model_0 = LinearRegressionModel()
print(list(model_0.parameters()))
print(model_0.state_dict())
```

```python
x_test, y_test
```

## Step 3: Train model. Fitting model to data'

```python
## set up a loss function
loss_fn = nn.L1Loss()

## set up an optimzier
optimizer = torch.optim.SGD(params=model_0.parameters(),
                           lr=0.001)
```

```python
num_epochs = 10
epoch_counts = []
train_losses = []
test_losses = []

for epoch in range(num_epochs):
  model_0.train()   # set model to training mode

  # forward pass
  y_pred = model_0(x_train)

  # calculate the training loss
  loss = loss_fn(y_pred, y_train)

  # Optimizer zero grad
  optimizer.zero_grad()

  # perform backpropogation
  loss.backward()

  # step the optimizer
  optimizer.step()

  model_0.eval()   # set model to evaluation mode

  # evaluation on test set
  with torch.inference_mode():
    test_pred = model_0(x_test)
    test_loss = loss_fn(test_pred, y_test)

  # save results
  epoch_counts.append(epoch)
  train_losses.append(loss.item())
  test_losses.append(test_loss.item())

  # print progress
  if epoch % 2 != 0:
    print(f"Epoch {epoch+1}/{num_epochs} | Train Loss: {loss.item():.4f} | Test Loss: {test_loss.item():.4f}")
```

## Prepare Data

```python
x = torch.arange(0, 20, 2, dtype=torch.float32)
x
```

```python
weight_1 = 3
bias_1 = 1
y = weight_1 * x + bias_1
y
```

```python
len(x)
```

```python
split_data = int(0.8 * len(x))
x_train, y_train = x[:split_data], y[:split_data]
x_test, y_test = x[split_data:], y[split_data:]
print(f" {x_train}\n {y_train}\n {x_test}\n {y_test}")
```

```python
## plot prediction
import matplotlib.pyplot as plt

def plot_prediction(train_data=x_train, train_labels=y_train,
                    test_data=x_test, test_labels=y_test,
                    prediction=None):
    plt.figure(figsize=(10,8))
    plt.scatter(train_data, train_labels, c="green", label="Training")
    plt.scatter(test_data, test_labels, c="blue", label="Testing")

    if prediction is not None:
        plt.scatter(test_data, prediction, c="red", label="Prediction")
        plt.legend(prop={"size":14})
    else:
        plt.legend(prop={"size":14})

    plt.show()
```

```python
plot_prediction(prediction=None)
```

## Build Model

```python
class LinearRegressionModel(nn.Module):
    def __init__(self) -> None:
        super().__init__()
        # define learnable parameters
        self.weight = nn.Parameter(torch.rand(1, requires_grad=True, dtype=torch.float))
        self.bias = nn.Parameter(torch.rand(1, requires_grad=True, dtype=torch.float))

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return self.weight * x + self.bias

model = LinearRegressionModel()
print(list(model.parameters()), model.state_dict())
```

## Fitting model to data

```python
# Loss and optimizer
loss_fn = nn.L1Loss()
optimizer = torch.optim.SGD(params=model.parameters(), lr=0.01)
epochs = 2000

# Training loop
epoch_counts = []
train_loss_value = []
test_loss_value = []

for epoch in range(epochs):
    model.train()
    train_preds = model(x_train)
    train_loss = loss_fn(train_preds, y_train)
    optimizer.zero_grad()
    train_loss.backward()
    optimizer.step()

    model.eval()
    with torch.inference_mode():
        test_preds = model(x_test)
        test_loss = loss_fn(test_preds, y_test)

    epoch_counts.append(epoch)
    train_loss_value.append(train_loss.item())
    test_loss_value.append(test_loss.item())

    if epoch % 50 == 0:
      print(f"Epoch {epoch+1}/{epochs}  | Train Loss: {train_loss.item():.4f}  | Test Loss: {test_loss.item():.4f}")
```

# Make prediction
