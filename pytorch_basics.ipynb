## Reshaping, view, stacking, squeezing, unsqueezing, permute tensors
* Reshaping - reshape an input to a defined shape
* View - return view of tensors of certain shape but keep the memory as the original tensor
* Stacking - combine multiple tesnors on each other vertical stack(vstack) or side by side(hstack)
* Squeezing - removes all `1` dimension from a tensor
* Unsqueezing - add a `1` dimension to a target tensor

## pyTorch workflow

```python
workflow = {1: "data(prepare and load  `1)",
            2: "Build Method",
            3: "Fitting model to data",
            4: "Making prediction",
            5: "saving and loading model"
            }
```

## Step 1 Data

-- spliting data into training and testing data set

```python
weight = 0.9
bias = 0.5

start = 0
end = 20
step = 0.5

x = torch.arange(start, end, step)
y = weight*x + bias
```

```python
x[:5], y[:5]
```

```python
len(x), len(y)
```

```python
split = int(0.8 * len(x))
x_train, y_train = x[:split], y[:split]
x_test, y_test = x[split:], y[split:]

len(x_train), len(y_train), len(x_test), len(y_test)
```

```python
## plot prediction
import matplotlib.pyplot as plt

def plot_prediction(train_data=x_train, train_labels=y_train,
                    test_data=x_test, test_labels=y_test,
                    prediction=None):
    plt.figure(figsize=(10,8))
    plt.scatter(train_data, train_labels, c="green", label="Training")
    plt.scatter(test_data, test_labels, c="blue", label="Testing")

    if prediction is not None:
        plt.scatter(test_data, prediction, c="red", label="Prediction")
        plt.legend(prop={"size":14})
    else:
        plt.legend(prop={"size":14})

    plt.show()
```

```python
plot_prediction()
```

## Step 2: Build Method

## Algorithms:
* Gradient descent
* Backpropogation

```python
# Create a linear regression model class
class LinearRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.weight = nn.Parameter(torch.rand(1, requires_grad=True, dtype=torch.float))
        self.bias = nn.Parameter(torch.rand(1, requires_grad=True, dtype=torch.float))

    # forward method defines the computation in the model
    def forward(self, x: torch.Tensor) -> torch.Tensor:   # x is input data
        return self.weight * x + self.bias
```

## pytorch mdoel building essentials
* torch.nn - cotains all of the building for computional graphs(a neural netwok can be considered a computional graph)
* torch.nm.Parameters  -  what parameters should  our model tey and often a pyTorch layer from torch.n will set these for us
* torch.nn.Module - The base class for all neural network modukes, if you subclass it, you should overwrite forward()
* torch.optim -this where the optimizers in pyTorch live, they help with gradient descent


```python
# Create a random seed
torch.manual_seed(42)

model_0 = LinearRegressionModel()
print(list(model_0.parameters()))
print(model_0.state_dict())
```

```python
x_test, y_test
```

## making prediction using `torch.inference_mode()`

```python
with torch.inference_mode():
    y_preds = model_0(x_test)

### this does similar thing, however, inference_mode is preffered
# with torch.no_grad():
#  y_preds = model_0(x_test)

print(y_preds)
```

```python
plot_prediction(prediction=y_preds)
```

## Step 3: Train model. Fitting model to data'

```python
## set up a loss function
loss_fn = nn.L1Loss()

## set up an optimzier
optimizer = torch.optim.SGD(params=model_0.parameters(),
                           lr=0.1)
```

```python
epochs = 1

for epoch in range(epochs):
  model_0.train()

  # forward pass
  y_pred = model_0(x_train)

  # calculate the loss
  loss = loss_fn(y_pred, y_train)

  # Optimzer zero grad
  optimizer.zero_grad()

  # perform backpropogation
  loss.backward()

  # step the optimizer
  optimizer.step()

  model_0.eval()
```
